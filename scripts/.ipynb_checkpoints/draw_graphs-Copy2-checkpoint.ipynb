```python
import os
import argparse
import pandas as pd
import logging
import altair as alt
from altair_saver import save
from vega_datasets import data
from datetime import datetime
```

```python
# let's load the data
data_gpu = pd.read_csv("~/Documents/BookCarla/Results/GPUs.csv", sep='\t', header=[0])
data_other = pd.read_csv("~/Documents/BookCarla/Results/OtherAppliences_short.csv", sep='\t', header=[0])
```

```python
def preproc_data(data):
    std_min=[avg - std/2 for avg, std in zip(data['avg'], data['std'])]
    std_max=[avg + std/2 for avg, std in zip(data['avg'], data['std'])]
    return pd.concat([pd.DataFrame({'x': data['Appliance'], 'std': std_min}), pd.DataFrame({'x': data['Appliance'], 'std': std_max})])
```

```python
gpu = preproc_data(data_gpu)
other = preproc_data(data_other)
print(data_gpu)
```

```python
errors = alt.Chart(gpu).mark_errorbar(extent='ci', color='#db646f').encode(
    y = alt.Y('x', sort=list(gpu['x']), title="", axis=alt.Axis(labels=False, ticks=False)),
    x=alt.X('std')
)
# Add points for life expectancy in 1955 & 2000

bars_gpu = alt.Chart(data_gpu).mark_point().encode(
    y=alt.Y('Appliance', axis=alt.Axis(labels=False, ticks=False), sort=list(gpu['x']), title=""),
    x=alt.X('avg')
)

bars_other = alt.Chart(data_other).mark_point().encode(
    y=alt.Y('Appliance', sort=list(other['x']), title=""),
    x=alt.X('avg')
)

# (errors + bars).resolve_scale(x='independent')
(bars_other + errors + bars_gpu).resolve_scale(y='independent')
```

```python
data = pd.read_csv("~/Documents/BookCarla/Results/GPUs_and_Others.csv", sep='\t', header=[0])
data_other = pd.read_csv("~/Documents/BookCarla/Results/OtherAppliences_short.csv", sep='\t', header=[0])
sorting_idx = data.sort_values('avg', ascending=False)['Appliance']
```

```python
bars_gpu = alt.Chart(data).mark_bar().encode(
    y=alt.Y('Appliance', sort=list(sorting_idx)),
    x=alt.X('avg', title='Average power draw (Watts)'),
    opacity = alt.Opacity('op', title='')
)

bars_other = alt.Chart(data_other).mark_bar(color="orange").encode(
    y=alt.Y('Appliance', sort=list(sorting_idx)),
    x=alt.X('avg'),
    opacity = alt.Opacity('op', title='')
)

(bars_gpu + bars_other).resolve_scale(opacity ='independent').configure_legend(orient='bottom').properties(width=300)
```

```python
data_2 = pd.read_csv("~/Documents/BookCarla/Results/Appliances_CO2.csv", sep='\t', header=[0])
print(data_2)
sorting_idx_2 = data_2.sort_values('CO2', ascending=False)['Appliance'].drop_duplicates()
print(sorting_idx_2)
```

```python
alt.Chart(data_2).mark_bar().encode(
    y=alt.Y('Country:N', title="", axis=alt.Axis(labels=False, ticks=False)),
    x=alt.X('CO2:Q', title='CO2 (kg) for a year'),
    row=alt.Row("Appliance", spacing=0, title="Appliance", sort=list(sorting_idx_2)),
    color = 'Country:N'
).configure_view(
    strokeWidth=0
).configure_headerRow(labelAngle=-0, labelAlign='left', labelAnchor='middle'
                     ).configure_legend(orient='bottom'
                                       ).properties(height=20, width=300)
```

```python
import altair as alt
import pandas as pd

data = pd.DataFrame([
    {"category":"A", "position":0, "value":0.1},
    {"category":"A", "position":1, "value":0.6},
    {"category":"A", "position":2, "value":0.9},
    {"category":"A", "position":3, "value":0.4},
    {"category":"B", "position":0, "value":0.7},
    {"category":"B", "position":1, "value":0.2},
    {"category":"B", "position":2, "value":1.1},
    {"category":"B", "position":3, "value":0.8},
    {"category":"C", "position":0, "value":0.6},
    {"category":"C", "position":1, "value":0.1},
    {"category":"C", "position":2, "value":0.2},
    {"category":"C", "position":3, "value":0.7}
])

text = alt.Chart(data).mark_text(dx=-10, color='white').encode(
    x=alt.X('value:Q', title=None),
    y=alt.Y('position:O', axis=None),
    text='value:Q'
)

bars = text.mark_bar().encode(
    color=alt.Color('position:O', legend=None, scale=alt.Scale(scheme='category20')),
)

(bars + text).facet(
    row='category:N'
).configure_header(
    title=None
)
```
